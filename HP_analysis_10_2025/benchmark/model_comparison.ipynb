```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Path to results folder
result_path = "results/models"

# Load all CSV files from the folder
all_results = []
for file in os.listdir(result_path):
    if file.endswith(".csv"):  # Ensure it's a CSV file
        file_path = os.path.join(result_path, file)
        df = pd.read_csv(file_path)
        df["Model"] = file.split("_")[1]  # Extract model name from filename
        all_results.append(df)

# Concatenate all dataframes
results_df = pd.concat(all_results, ignore_index=True)

print(results_df)

# Plot performance metrics for different feature sets
sns.set_style("whitegrid")
fig, axes = plt.subplots(1, 3, figsize=(10, 4), gridspec_kw={'width_ratios': [1, 1, 0.3]})

for idx, k in enumerate([10, 1000]):
    subset_df = results_df[results_df["Features (k)"] == k]
    
    # Define common y-axis range
    y_min = 0.45
    y_max = 0.75
    
    # OvR Macro AUC (Bar Chart)
    ax = axes[idx]
    sns.barplot(x="Model", y="OvR Macro AUC", hue="Model", data=subset_df, palette="viridis", ax=ax)
    ax.set_ylim(y_min, y_max)
    ax.set_title(f"OvR Macro AUC (Features = {k})")
    ax.set_xlabel("Model")
    ax.set_ylabel("AUC Score")
    ax.get_legend().remove()  # Remove individual legends

# Create a single legend in the extra subplot
handles, labels = axes[0].get_legend_handles_labels()
axes[2].legend(handles, labels, loc='center')
axes[2].axis('off')  # Hide the subplot axes

plt.tight_layout()
plt.show()
```
